```python
# filtering sales to only RMA geos
sales_RMA = sales[np.array(sales.Geography.apply(lambda x: re.search("-RMA", x))) != None]
sales_RMA.head(10)
```

```python
# Fixing Time column
# is it always ending in 4 weeks?
print(sales_RMA.shape == 
      sales_RMA[np.array(sales_RMA.Time.apply(lambda x: re.search("4 Weeks Ending", x))) != None].shape) #yes

sales_RMA['end_time'] = sales_RMA.Time.apply(lambda t: re.split(" ", t)[-1])
sales_RMA.end_time = sales_RMA.end_time.apply(lambda x: datetime.datetime.strptime(x, "%m-%d-%y"))
sales_RMA['start_time'] = sales_RMA.end_time.apply(lambda x: x - datetime.timedelta(weeks = 4))
```
